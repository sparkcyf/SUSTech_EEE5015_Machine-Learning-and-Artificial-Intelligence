This file is intend to split the handwrite data, and generate the data for training and testing.

```python
import cv2
import numpy as np
```

```python
# read image
img = cv2.imread('mnist3.png', cv2.IMREAD_GRAYSCALE)
```

```python
# resize image to 224x280
img = cv2.resize(img, (224, 280))
```

```python
# plot image
import matplotlib.pyplot as plt
plt.imshow(img, cmap='gray')
```

```python
# plot img[i*28:(i+1)*28, j*28:(j+1)*28]
i = 8
j = 0
plt.imshow(img[i*28:(i+1)*28, j*28:(j+1)*28], cmap='gray')
```

```python
# split image into 28x28 images
imgs = np.zeros((80, 28, 28))
count = 0
for i in range(10):
    for j in range(8):
        imgs[count] = img[i*28:(i+1)*28, j*28:(j+1)*28]
        count += 1
```

```python
# reverse image
imgs = 255 - imgs
```

```python
# clean up border line
imgs[:, 0:2, :] = 0
imgs[:, 26:28, :] = 0
imgs[:, :, 0:2] = 0
imgs[:, :, 26:28] = 0
```

```python
# plot images
fig, axs = plt.subplots(10, 8, figsize=(8, 10))
count = 0
for i in range(10):
    for j in range(8):
        axs[i, j].imshow(imgs[count], cmap='gray')
        count += 1
        axs[i, j].axis('off')
```

```python
# save images to npy file
np.save('spark_handwrite_test5.npy', imgs)
```
